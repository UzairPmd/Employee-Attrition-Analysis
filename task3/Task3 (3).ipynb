```python
import numpy as np
def get_random():
    return np.random.randint(1, 101)
r = get_random()
n = int(input("Guess the Number (1-100): "))
if n > r:
    print(f"Your guess is too high! The number was {r}.")
elif n < r:
    print(f"Your guess is too low! The number was {r}.")
else:
    print("You guessed it right!")
```

```python
import numpy as np
def get_random():
    return np.random.randint(1, 101) 
r = get_random()
print("Guess the Number (between 1 and 100):")
while True:
    n = int(input("Enter your guess: "))
    if n > r:
        print("Too high! Try again.")
    elif n < r:
        print("Too low! Try again.")
    else:
        print("You guessed it right!")
        break  
```
